# ApiPlaygroundPy Notebook

This notebook demonstrates basic usage of the **ApiPlaygroundPy** modules for calling an API and doing simple data processing.


```python
import sys, os

# Ensure project root is on sys.path when running this notebook directly
project_root = os.path.abspath(os.path.join(os.getcwd()))
if project_root not in sys.path:
    sys.path.append(project_root)

from src.api_playground import ApiPlayground
from src.data_processing import to_dataframe, select_columns, filter_rows
from src.utils.logger import get_logger
```

```python
logger = get_logger(__name__)
logger.info("Starting API playground notebook...")

client = ApiPlayground("https://jsonplaceholder.typicode.com")
posts = client.get("/posts")
len(posts)
```

```python
# Convert to DataFrame and inspect
df = to_dataframe(posts)
df.head()
```

```python
# Select some columns
df_small = select_columns(df, ['userId', 'id', 'title'])
df_small.head()
```

```python
# Filter rows for a specific user
df_user_1 = filter_rows(df_small, lambda d: d['userId'] == 1)
df_user_1.head()
```
